```python
import pymortar
# connect client to Mortar frontend server
client = pymortar.Client("https://beta-api.mortardata.org")
```

```python
# query thermostat
query = """SELECT ?setpoint ?zone WHERE {{
    ?setpoint a brick:Zone_Air_Temperature_Setpoint .
    ?vav brick:hasPoint ?setpoint .
    ?vav brick:feeds ?zone .
}}"""

# query = """SELECT * WHERE {{
#     ?rm a brick:Room .
#     ?type brick:subClassOf* ?rm.
# }}"""
```

### Across 1611 zones in Mortar portfolio
- more than 50% zones have less than 10 adjustments of thermostat setpoints
- more than 80% zones have less than 100 adjustments of thermostat setpoints

```python
pd.set_option('display.max_rows', 100)
metadata_all = client.sparql(query, sites=qualify_resp.sites[1:])
metadata_all
```

```python
import re
res_data = []
for idxi, row in metadata_all.iterrows():
    res_row = {}
    n_count = 0.0
    md = row.setpoint
    data_sensor = client.data_uris([md])
    data = data_sensor.data
    # get a pandas dataframe between start date and end date of the data
    sd_ns = pd.to_datetime(sd, unit='ns', utc=True)
    ed_ns = pd.to_datetime(ed, unit='ns', utc=True)
    df = data[(data['time'] >= sd_ns) & (data['time'] <= ed_ns)]
    # parse the hour and weekday info and add it as a column
    df['hr'] = pd.to_datetime(df['time']).dt.hour
    df['wk'] = pd.to_datetime(df['time']).dt.dayofweek
    df['mo'] = pd.to_datetime(df['time']).dt.month
    # create occupied df by normal office hours and by weekdays
    df_occ = df[(df['hr'] >= sh) & (df['hr'] < eh) &
                (df['wk'] >= 0) & (df['wk'] <= 4)]
    # get the difference between current row and previous row
    df_count = df_occ['value'].diff().value_counts()
#     print(df_count)
#     # calculate all adjustments
#     n_count = df_count.sum()-df_count[0]
    # calcualte adjustments that are between 1 to 10 degrees
    for adj in df_count.keys():
        n_count = n_count + df_count[adj] if abs(adj) in range(11)[1:] else n_count
    # store results into a dict and append dict as a row in the list
    res_row['zone'] = re.split('[#]', row.zone)[-1]
    res_row['site'] = row.site
    res_row['index'] = n_count
#     print(res_row)
    res_data += [res_row]
res_df = pd.DataFrame(res_data)
res_df.sort_values(['index'], ascending=[False])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

data = res_df['index']
# set bin width as 1
# n_bins = 10
binwidth = 10
n_bins = np.arange(min(data), max(data) + binwidth, binwidth)

plt.figure(figsize=(20, 10), dpi=80)
plt.hist(data, bins=n_bins, weights=np.ones(len(data))/len(data))
plt.xlabel('number of adjustment during one year')
plt.ylabel('percentage of zones')
plt.title('Thermostat adjustment frequency distribution across the whole Mortar')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.show()
```
